### 1. 스위스롤 예시

- 참고: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_swiss_roll.html
- 참고: https://scikit-learn.org/stable/modules/generated/sklearn.manifold.LocallyLinearEmbedding.html

```python
import sklearn
from sklearn import datasets, manifold

import matplotlib.pyplot as plt
```

```python
X, color = sklearn.datasets.make_swiss_roll(n_samples=3000, random_state=0)
```

```python
X
```

```python
color
```

```python
print("Computing LLE embedding")
X_r, err = manifold.locally_linear_embedding(X, n_neighbors=20, n_components=2)
print("Done. Reconstruction error: %g" % err)
```

```python
fig = plt.figure(figsize=(8,14))

ax = fig.add_subplot(211, projection="3d")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.set_title("Original data")

ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis("tight")
plt.xticks([]), plt.yticks([])
plt.title("Projected data By LLE")

plt.show()
```

```python
print(X[:1], X_r[:1])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_p = pca.fit_transform(X)
```

```python
plt.scatter(X_p[:, 0], X_p[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis("tight")
plt.xticks([]), plt.yticks([])
plt.title("Projected data By PCA")

plt.show()
```

### 2. S-Curve 예시

- 참고: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_s_curve.html

```python
print("Computing LLE embedding")
Y_r, err = manifold.locally_linear_embedding(Y, n_neighbors=14, n_components=2)
print("Done. Reconstruction error: %g" % err)
```

```python
fig = plt.figure(figsize=(8,14))

ax = fig.add_subplot(211, projection="3d")
ax.scatter(Y[:, 0], Y[:, 1], Y[:, 2], c=color, cmap=plt.cm.Spectral)
ax.set_title("Original data")

ax = fig.add_subplot(212)
ax.scatter(Y_r[:, 0], Y_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis("tight")
plt.xticks([]), plt.yticks([])
plt.title("Projected data")

plt.show()
```

```python
print(Y[:1], Y_r[:1])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
Y_p = pca.fit_transform(Y)
```

```python
plt.scatter(Y_p[:, 0], Y_p[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis("tight")
plt.xticks([]), plt.yticks([])
plt.title("Projected data By PCA")

plt.show()
```
